# bivariate analysis

as we want to predict price our main focus on selling price in bivariate analysis

```python
plt.figure(figsize=(10,10))
sns.boxplot(x=df["company"],y=df['selling_price'])
plt.xticks(rotation=90)
plt.ticklabel_format(style="plain",axis='Y')
```

column can be useful as selling price median values are varying much across car names

```python
sns.scatterplot(x=df["year_of_purchase"],y=df["selling_price"])
plt.ticklabel_format(style="plain",axis='Y')
plt.ticklabel_format(style="plain",axis='Y')
```

as year increses selling price of car is more
old car getting less price (logically also true)

```python
sns.scatterplot(x=df["km_driven"],y=df["selling_price"])
plt.ticklabel_format(style="plain",axis='Y')
plt.ticklabel_format(style="plain",axis='x')
```

not able to find clear relationship

```python
plt.figure(figsize=(10,10))
#plt.subplots(figsize=(10,10))
sns.boxplot(x=df["fuel_type"],y=df['selling_price'])
plt.xticks(rotation=90)
plt.ticklabel_format(style="plain",axis='Y')
```
